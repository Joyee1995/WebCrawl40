### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
soup
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
title=[]
titles=[]
for i in soup.select('div.part_list_2 h3'):
        titles.append(i.text)
#     title.append(i.find(class_="date").text)
#     title.append(i.find_all('a')[-1].text)
#     titles.append(title)
    
        

   
```

```python
import re
for title in titles:
#     print(title)
    time=re.findall('[0-9]{2}:[0-9]{2}',title)
    for time1 in time:
        datesplit=title.split(time1)
        print(datesplit[0],time1,datesplit[1])
```
